```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# upload the data
df1=pd.read_csv("benin_clean.csv")
df2=pd.read_csv("togo_clean.csv")
df3=pd.read_csv("sierraleone_clean.csv")

df=[df1,df2,df3]
```

```python
df1[df1.isnull().any(axis=1)]
df1=df1.dropna()
```

1. Metric Comparision

```python
#prepare the data
col=["GHI,DNI,DHI"]
name=["benin","togo","sierraleone"]
GHI_column=[df1["GHI"],df2["GHI"],df3["GHI"]]
DNI_column=[df1["DNI"],df2["DNI"],df3["DNI"]]
DHI_column=[df1["DHI"],df2["DHI"],df3["DHI"]]
```

```python
name=["benin","togo","sierraleone"]
colors = ['red', 'green', 'blue']

#Boxplot for GHI
fig, ax = plt.subplots()
ax.set_ylabel('GHI')
bplot = ax.boxplot(GHI_column, patch_artist=True, tick_labels=name, showfliers=False)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.title("boxplot for GHI")
plt.show()

#Boxplot for DNI
fig, ax = plt.subplots()
ax.set_ylabel('DNI')
bplot = ax.boxplot(DNI_column, patch_artist=True, tick_labels=name, showfliers=False)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.title("boxplot for DNI")
plt.show()


#Boxplot for DHI
fig, ax = plt.subplots()
ax.set_ylabel('DHI')
bplot = ax.boxplot(DHI_column, patch_artist=True, tick_labels=name, showfliers=False)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.title("boxplot for DHI")
plt.show()
```

```python
#Summary Table
s1=df1.agg(
    {
        "GHI": ["mean", "median", "std"],
        "DNI": ["mean", "median", "std"],
        "DHI": ["mean", "median", "std"]
    })
s2=df2.agg(
    {
        "GHI": ["mean", "median", "std"],
        "DNI": ["mean", "median", "std"],
        "DHI": ["mean", "median", "std"]
    })
s3=df3.agg(
    {
        "GHI": ["mean", "median", "std"],
        "DNI": ["mean", "median", "std"],
        "DHI": ["mean", "median", "std"]
    })
combined_df=pd.concat([s1,s2,s3],axis=1,keys=["benin","togo","sierraleone"])
print(combined_df)
```

2. Key Observations

+ Benin has the highest average GHI and DNI while the three countries have similar values of DHI.

+ Sierraleone has the lowest variations in GHI and DNI values while Benin has the highest.

+ The DHI values of each country has similar variations.
